```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randint(1, 10, size=(10, 15)), 
                  columns=['col1', 'col2', 'col3', 'col4', 'col5', 
                           'col6', 'col7', 'col8', 'col9', 'col10',
                           'col11', 'col12', 'col13', 'col14', 'col15']) 
df
```

```python
for col in df:
    df[col] = df[col].replace(1, 1111)
    df[col] = df[col].replace(4, 4444)
    df[col] = df[col].replace(9, 9999)
df
```

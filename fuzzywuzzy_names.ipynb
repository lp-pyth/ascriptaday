```python
import wikipedia as wiki
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
import spacy
nlp = spacy.load("en_core_web_sm") 
```

```python
def get_friends(name):
    content = wiki.page(name).content
    doc = nlp(content)
    ents_list = []
    for ent in doc.ents:
        if ent.label_=="PERSON" and ent.text not in ents_list and len(ent.text.split()) > 1:
            ents_list.append(ent.text)
    return ents_list
```

```python
level_1 = get_friends(name)
```

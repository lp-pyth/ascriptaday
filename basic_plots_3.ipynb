```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
df = pd.read_csv("rent.asc", delim_whitespace=True)
df = df.astype(np.float64)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(22, 16))
for xcol, ax in zip(df.columns[1:-1], [x for v in axes for x in v]):
    df.plot.scatter(x='nm', y=xcol, ax=ax, alpha=0.5, color='orange')
    ax.set_facecolor("#3F495A")
```

```python
sns.set_theme(style="darkgrid", palette='deep')
corr = df.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)
```

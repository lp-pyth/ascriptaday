```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame(np.random.randint(1, 20, size=(6, 3)), 
                  columns=['col1', 'col2', 'col3'])
df['col4']=df['col2']+df['col3']
for i in range(len(df)):
    if (i % 2) == 0:
        df.loc[i, 'col1']='A'
    else:
        df.loc[i, 'col1']='B'
```

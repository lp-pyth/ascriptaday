```python
import pandas as pd
import numpy as np 
df = pd.DataFrame(dict(A=['ada', 'eth', 'ada', 'eth',
                          'eth', 'ada', 'ada'],
                       B=['usd','usdt', 'usd', 'usdt',
                          'usd', 'usdt', 'usdt'],
                       col_1=[1,2,5,4,3,2,7],
                       col_2=[8,6,3,4,2,3,3]))
table = df.groupby(['A','B']).agg({'col_1': ['max', np.mean],
                                 'col_2': ['sum','min','count']})
table
```

```python
df = pd.DataFrame(dict(A=['A','A','A','A','A','B','B','B','B'],
                       B=[1,1,1,2,2,1,1,1,2],
                       C=['ada','eth','ada','ltc','ltc','etc','ltc','eth','etc']))
df
```

```python
table = df[['A', 'B', 'C']].drop_duplicates()\
                         .groupby(['A','B'])['C']\
                         .apply(list)\
                         .reset_index()# list to string (separated by commas) 
table['C'] = table.apply(lambda x: (','.join([str(s) for s in x['C']])), axis = 1)
table
```

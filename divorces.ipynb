```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
table_MN = pd.read_html('https://www.scheidung.org/scheidungsstatistik')
df = table_MN[0]    
df = df.set_index("Jahr")
```

```python
df=df.rename(columns = {'Scheid­ungen':'divorces', 
                        'Ehe­schlieÃ­ungen':'marriages', 
                        'jÃ¤hr­liche Scheidungs­rate (gerundet in %)':'divorcerate' })
df.drop(df.tail(2).index,inplace=True)
df.loc["2016", "marriages"] = 410000
df['marriages'] = df['marriages'].str.replace('\.','').astype('float')
df['divorces'] = df['divorces'].str.replace('\.','').astype('float')
df['divorcerate'] = df['divorcerate'].astype('float')/10
df.loc["2016", "marriages"] = 410000
```

```python
labels = ['divorces', 'marriages', 'divorcerate']
dat = df[['divorces', 'marriages',   'divorcerate']]

axes = df.plot(kind='line', figsize=(20,4), title='Marriages and Divorces Germany', subplots=True, layout=(1,3))

axes[0,0].set_facecolor('#3F495A')
axes[0,1].set_facecolor('#3F495A')
axes[0,2].set_facecolor('#3F495A')
```

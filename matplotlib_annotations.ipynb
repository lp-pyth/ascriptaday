```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
x = [x for x in range(10)]
y = [2, 8, 6, 1, 3, 2, 7, 1, 8, 6]
text = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"]

ax.set_facecolor("#3F495A") 
ax.set_xlabel('X Values',size=10, color="#C9E2FA")   
ax.set_ylabel('Y Values',size=10, color="#C9E2FA") 
ax.scatter(x, y, color='orange', alpha=1)

for i in range(len(x)):
    plt.annotate(text[i], (x[i], y[i] - 0.35), color="#C9E2FA")

ax.set_xlim((-1, 10))
ax.set_ylim((0, 10))
```

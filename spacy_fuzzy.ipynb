```python
import spacy
from spacy import displacy
import en_core_web_sm
nlp = en_core_web_sm.load()
doc = nlp(text)
#for ent in doc.ents:
#    print(ent.text, ent.start_char, ent.end_char, ent.label_)
```

```python
person_list = []
for entity in doc.ents:
    if entity.label_ in ["PERSON"]:
        print(entity.text + ' - ' + entity.label_ + ' - ' + str(spacy.explain(entity.label_)))
        person_list.append(str(entity.text))
```

```python
import fuzzywuzzy
from fuzzywuzzy import process, fuzz
person_fuzzied = []
for p1 in person_list:
    for p2 in person_list:
        ratio = fuzz.ratio(p1.lower(), p2.lower())
        print(p1, p2, 'Similarity score: {}'.format(ratio))
```

```python
#Prepare colors for displacy (to show text and tags)
import random 
def get_entity_options(random_colors=False):
    def color_generator(number_of_colors):
        color = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
                 for i in range(number_of_colors)]
        return color
    entities = ["PERSON", "ORG"]
    #entities = ["PER", "LOC", "ORG", "CHEBI"]
    colors = {"ENT":"#E8DAEF"}    
    if random_colors:
        color = color_generator(len(entities))
        for i in range(len(entities)):
            colors[entities[i]] = color[i]
    else:
        entities_cat_1 = {"PERSON":"darkorange"}
        entities_cat_2 = {"ORG":"violett"}
        #entities_cat_3 = {"ORG":"blue"}
        #entities_cats = [entities_cat_1, entities_cat_2, entities_cat_3]
        entities_cats = [entities_cat_1, entities_cat_2]
        for item in entities_cats:
            colors = {**colors, **item}    
    options = {"ents": entities, "colors": colors}
    return options
```

```python
options = get_entity_options()
displacy.render(doc, style='ent', options=options)
```

```python
import pandas as pd
```

```python
data = [["the cat and the dog"], 
        ["the panda and the elephant"], 
        ["the dog and the cat"], 
        ["two mice, seven cows"], 
        ["two dogs and the cat"], 
        ["a cat and four ducks, and seventeen chicken"]]

df = pd.DataFrame(data) 
df.columns =['Text']
df
```

```python
#Duplicates
duplicates_list = []
for r in range(0, len(df)-1):
    words1 = set(df["Text"][r].split())
    for s in range(r+1, len(df)):
        temp_list=[]
        words2 = set(df["Text"][s].split())
        duplicates = words1.intersection(words2)
        uniques = words1.difference(words2).union(words2.difference(words1))
        overlap = len(duplicates)/(len(duplicates)+len(uniques))
        temp_list.append(df.index[r])
        temp_list.append(df.index[s])
        temp_list.append("%.2f" % overlap)
        duplicates_list.append(temp_list)
df_dup = pd.DataFrame(duplicates_list) 
df_dup.columns = [["Text1", "Text2", "Overlap"]]
df_dup
```

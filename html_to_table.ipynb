```python
import pandas as pd
```

```python
html_url = "https://en.wikipedia.org/wiki/South_Park"
tables = pd.read_html(html_url)
len(tables)
```

```python
for i in range(16):
    print(i, '--------------------------------------------------------')
    print(tables[i])
```

```python
df = tables[1]
df.head(2)
```

```python
df.drop(df.columns[[0, 3]], axis=1, inplace=True)
df.columns = ['Season', 'Episodes', 'First Aired', 'Last Aired']
df.set_index('Season', inplace=True)
df['First Aired'] = pd.to_datetime(df['First Aired'])
df['Last Aired'] = pd.to_datetime(df['Last Aired'])
df.head(2)
```

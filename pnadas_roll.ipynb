```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame(np.random.randint(1, 50, size=(100, 5)), 
                  columns=['col1', 'col2', 'col3', 'col4', 'col5']) 
```

```python
df.head()
```

```python
days = 7
df_mean_7 = df.rolling(days).sum()/days
days = 14
df_mean_14 = df.rolling(days).sum()/days
```

```python
df.col1.plot(figsize=(22,12), color="grey")
df_mean_7.col1.plot()
df_mean_14.col1.plot()
```
